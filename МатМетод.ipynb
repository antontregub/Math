```
from scipy.signal import *
from pylab import title, grid, xlim, matrix
import matplotlib.pyplot as plt
b=[1, -1, 1, -1]
a = [1, 0,  1, 4]
z,p,k = tf2zpk(b,a)
print(z)
print(p)
print(k)
w, H = freqresp([z,p,k])
plt.figure()
plt.grid()
plt.plot(H.real, H.imag, "b")
title('Godograf')
plt.show()
```

```

T, yout = impulse([b,a])
plt.figure()
plt.grid()
plt.plot(yout, "r")
title('Impulse')
plt.show()
```

```
T, yout = step([b,a])
plt.figure()
plt.grid()
plt.plot(yout, "r")
title('Step')
plt.show()
```

```
w,mag,phase = bode([b,a])
plt.figure()
plt.grid()
plt.semilogx(w, mag)    # Bode magnitude plot
title('AChH')
plt.figure()
plt.grid()
plt.semilogx(w, phase)  # Bode phase plot
title('FChH')
plt.show()
```
